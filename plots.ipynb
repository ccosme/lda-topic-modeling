# Import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set_style('white')
```

# Plot trends

```python
# Read data
loc = '/data/trends.xlsx'
data = pd.read_excel(loc)
data
```

```python
plt.plot(data.Year, data.Topic_0, color='dodgerblue', label='Topic 1')
plt.plot(data.Year, data.Topic_1, color='darkgoldenrod', label='Topic 2')
plt.plot(data.Year, data.Topic_2, color='mediumorchid', label='Topic 3')
plt.plot(data.Year, data.Topic_3, color='mediumvioletred', label='Topic 4')
plt.plot(data.Year, data.Topic_4, color='orange', label='Topic 5')
plt.plot(data.Year, data.Topic_5, color='blueviolet', label='Topic 6')
plt.plot(data.Year, data.Topic_6, color='deeppink', label='Topic 7')
plt.plot(data.Year, data.Topic_7, color='deepskyblue', label='Topic 8')
plt.plot(data.Year, data.Topic_8, color='brown', label='Topic 9')
plt.plot(data.Year, data.Topic_9, color='royalblue', label='Topic 10')
plt.plot(data.Year, data.Topic_10, color='limegreen', label='Topic 11')
plt.plot(data.Year, data.Topic_11, color='darkorange', label='Topic 12')
plt.plot(data.Year, data.Topic_12, color='forestgreen', label='Topic 13')
plt.xticks(data.Year)
plt.legend(loc='upper left')
plt.ylim(0,1800)
plt.show()
```

```python
# upward trend
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 9
plt.plot(data.Year, data.Topic_0, color='dodgerblue', label='Topic 1') # Many-body methods
plt.plot(data.Year, data.Topic_1, color='darkgoldenrod', label='Topic 2') # Quantum magnetism
plt.plot(data.Year, data.Topic_5, color='blueviolet', label='Topic 6') # Structural properties of materials
plt.plot(data.Year, data.Topic_6, color='deeppink', label='Topic 7') # Amorphous solids
plt.plot(data.Year, data.Topic_7, color='deepskyblue', label='Topic 8') # Topological physics
plt.plot(data.Year, data.Topic_10, color='limegreen', label='Topic 11') # Graphene
plt.xticks(data.Year)
plt.ylim(0,1800)
plt.legend(loc='lower right')
plt.xlabel('Year')
plt.ylabel('Number of published articles')
plt.tick_params(bottom=True, left=True)
plt.show()
```

```python
# downward trend
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 9
plt.plot(data.Year, data.Topic_3, label='Topic 4') # Bose-Einstein condensates
plt.plot(data.Year, data.Topic_8, color='brown', label='Topic 9') # Quantum electronic devices 
plt.xticks(data.Year)
plt.ylim(0,1800)
plt.legend(loc='lower right')
plt.xlabel('Year')
plt.ylabel('Number of published articles')
plt.tick_params(bottom=True, left=True)
plt.show()
```

```python
# steady trend
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 9
plt.plot(data.Year, data.Topic_2, color='mediumorchid', label='Topic 3') # Quantum magnetism
plt.plot(data.Year, data.Topic_4, color='orange', label='Topic 5') # Dynamical phases of matter
plt.plot(data.Year, data.Topic_9, color='royalblue', label='Topic 10') # Polymer/Soft condensed matter physics
plt.plot(data.Year, data.Topic_12, color='forestgreen', label='Topic 13') # Superconductivity
plt.xticks(data.Year)
plt.ylim(0,1800)
plt.legend(loc='lower right')
plt.xlabel('Year')
plt.ylabel('Number of published articles')
plt.tick_params(bottom=True, left=True)
plt.show()
```

```python
# create subplots for all three trends

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=True)

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 9

fig.set_figheight(4)
fig.set_figwidth(15)

# upward trend
ax1.plot(data.Year, data.Topic_0, color='tomato', label='Topic 1') # Many-body methods
ax1.plot(data.Year, data.Topic_1, color='limegreen', label='Topic 2', linestyle='--') # Quantum magnetism
ax1.plot(data.Year, data.Topic_5, color='deeppink', label='Topic 6') # Structural properties of materials
ax1.plot(data.Year, data.Topic_6, color='blueviolet', label='Topic 7', linestyle='--') # Amorphous solids
ax1.plot(data.Year, data.Topic_7, color='orange', label='Topic 8') # Topological physics
ax1.plot(data.Year, data.Topic_9, color='royalblue', label='Topic 10', linestyle='--') # Polymer/Soft condensed matter physics
ax1.plot(data.Year, data.Topic_10, color='darkgreen', label='Topic 11') # Graphene
ax1.set_xticks(data.Year)
ax1.set_ylim(0,1800)
ax1.legend(loc='lower right')
ax1.set_title('Upward trend', fontsize=12)
ax1.set_ylabel('Number of published articles', fontsize=12)
ax1.tick_params(bottom=True, left=True)

# downward trend
ax2.plot(data.Year, data.Topic_3, label='Topic 4') # Bose-Einstein condensates
ax2.plot(data.Year, data.Topic_8, color='brown', label='Topic 9') # Quantum electronic devices 
ax2.set_title('Downward trend', fontsize=12)
ax2.set_xticks(data.Year)
ax2.set_ylim(0,1800)
ax2.legend(loc='lower right')
ax2.set_xlabel('Year', fontsize=12)
ax2.tick_params(bottom=True)

# steady trend
ax3.plot(data.Year, data.Topic_2, color='mediumorchid', label='Topic 3') # Quantum magnetism
ax3.plot(data.Year, data.Topic_4, color='orange', label='Topic 5') # Dynamical phases of matter
ax3.plot(data.Year, data.Topic_12, color='forestgreen', label='Topic 13') # Superconductivity
ax3.set_title('Steady trend', fontsize=12)
ax3.set_xticks(data.Year)
ax3.set_ylim(0,1800)
ax3.legend(loc='lower right')
ax3.tick_params(bottom=True)


plt.savefig('fig_trend_subplots.pdf')
plt.show()
```
